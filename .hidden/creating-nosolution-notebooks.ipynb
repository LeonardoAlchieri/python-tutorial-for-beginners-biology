```python
import json
from pathlib import Path



for path in Path('../notebooks_with_solutions').glob('**.ipynb'):
    # Load notebook
    with open(path, 'r') as f:
        notebook = json.load(f)

    # Filter out hidden cells
    notebook['cells'] = [cell for cell in notebook['cells'] 
                        if not cell.get('metadata', {}).get('slideshow', {}).get("slide_type", False) == 'skip']

    # Save filtered notebook
    new_path = Path('..') / f"{path.stem}-CLEAN{path.suffix}"
    with open(new_path, 'w', encoding='utf-8') as f:
        json.dump(notebook, f, indent=1)
```
